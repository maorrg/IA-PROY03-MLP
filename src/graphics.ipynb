```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RESULTS_DIRECTORY = 'data/output/results/'
OUTPUT_DIRECTORY = 'data/output/graphics/'
```

```python
headers = ['precision','recall','f1_score','epoch', 'activation_function']
data = pd.read_csv(RESULTS_DIRECTORY + 'g1.csv', sep=",", names=headers)
```

```python
fig, axs = plt.subplots(1, 3)
axs[0].set_title("Precision")
axs[0].plot(data['epoch'],data['precision'], label="Sigmoide", color="green")
axs[0].plot(data['epoch'],data['recall'], label="Tanh", color="orange")
axs[0].plot(data['epoch'],data['f1_score'], label="RELU", color="red")
axs[0].set_box_aspect(1)

axs[1].set_title("Recall")
axs[1].plot(data['epoch'],data['recall'], color="orange")
axs[1].set_box_aspect(1)

axs[2].set_title("F1 Score")
axs[2].plot(data['epoch'],data['f1_score'], color="red")
axs[2].set_box_aspect(1)

# plt.xlabel("Epochs")
# plt.ylabel("Train value")
fig.legend(loc="center right", bbox_to_anchor = (1.15, 0.68))
fig.suptitle('Main title')
fig.subplots_adjust(top=1.2)
plt.show()
```

```python
filtered_sigmoide = data[data['activation_function'] == 'sigmoide']
filtered_relu = data[data['activation_function'] == 'relu']
filtered_tanh = data[data['activation_function'] == 'tanh']

plt.plot(filtered_sigmoide['epoch'],filtered_sigmoide['precision'], label="Sigmoide", color="green")
plt.plot(filtered_relu['epoch'],filtered_relu['precision'], label="RELU", color="orange")
plt.plot(filtered_tanh['epoch'],filtered_tanh['precision'], label="Tanh", color="red")

plt.title("Epochs vs. Valor de Precisión en la etapa de \n Training para cada función de activación")
plt.legend(loc="upper left") #bbox_to_anchor = (1.28, 0.5)
plt.xlabel("Epochs")
plt.ylabel("Valor de Precisión")
plt.show()
```
