```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RESULTS_DIRECTORY = 'data/output/results/'
OUTPUT_DIRECTORY = 'data/output/graphics/'
```

```python
headers = ['precision','recall','f1_score','epoch', 'activation_function']
data = pd.read_csv(RESULTS_DIRECTORY + 'g1.csv', sep=",", names=headers)
```

```python
def generate_activation_functions_graphics(metric):
    filtered_sigmoide = data[data['activation_function'] == 'sigmoide']
    filtered_relu = data[data['activation_function'] == 'relu']
    filtered_tanh = data[data['activation_function'] == 'tanh']

    plt.plot(filtered_sigmoide['epoch'],filtered_sigmoide[metric], label="Sigmoide", color="green")
    plt.plot(filtered_relu['epoch'],filtered_relu[metric], label="RELU", color="orange")
    plt.plot(filtered_tanh['epoch'],filtered_tanh[metric], label="Tanh", color="red")

    plt.title(f"Epochs vs. Valor de {metric} en la etapa de \n Training para cada función de activación")
    plt.legend(loc="upper left") #bbox_to_anchor = (1.28, 0.5)
    plt.xlabel("Epochs")
    plt.ylabel(f"Valor de {metric}")
    plt.show()
```

```python
metrics = ['precision','recall','f1_score']

for metric in metrics:
    generate_activation_functions_graphics(metric)
```
